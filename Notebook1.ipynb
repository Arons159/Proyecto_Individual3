```python
# Quitamos las columnas que tienen muchos datos faltantes y generan ruido a nuesto dataset
def drop_col(df):
    df.drop(columns=['flight_no','cn_ln','PASAJEROS A BORDO','ground', 'registration'])

drop_col(df_acc)
df_acc['Ruta2'] = df_acc['route']
```

```python
# Nos sirve para cambiar el formato de la fecha
def date_convert(date_to_convert):
     return datetime.datetime.strptime(date_to_convert, "%B %d, %Y").strftime('%m/%d/%Y')
# Ejecutamos la funcion
df_acc['fecha'] = df_acc['fecha'].apply(date_convert)
```

```python
# Se cambian los nombres de las variables 
def transform_main(df):
    main_df = df
    main_df.rename(columns= {'Unnamed': 'Numero',
        'fecha': 'Fecha_accidente',
        'HORA declarada': 'Colision_hora',
        'Ruta': 'Origen',
        'OperadOR': 'Operador',
        'route': 'Tipo_vuelo',
        'ac_type': 'Tipo_nave',
        'registration': 'Codigo',
        'all_aboard': 'Pasajeros',
        'PASAJEROS A BORDO': 'Pasajeros_a_bordo',
        'crew_aboard': 'Tripulacion',
        'cantidad de fallecidos': 'Numero_de_fallecidos',
        'passenger_fatalities': 'Pasajeros_fallecidos',
        'crew_fatalities': 'Tripulantes_fallecidos',
        'summary': 'Descripcion',
        'ground': 'Fallecidos_suelo',
        'Ruta2': 'Ruta'
        }, inplace= True)

transform_main(df_acc)    
```

```python
# Se itera para cambiar los valores '?' por 'Desconocido'
for col in df_acc.columns:
    if df_acc[col].dtype == object:
        df_acc = df_acc.replace('?', 'Desconocido')
```

```python
# Se cambia el formato de hora
df_acc['Colision_hora'] =  df_acc['Colision_hora'].replace('Desconocido', None)
df_acc['Colision_hora'] = df_acc['Colision_hora'].str.replace(':','')
df_acc['Colision_hora'] = df_acc['Colision_hora'].str.replace(r'\D+','',regex=True)
df_acc['Colision_hora'] = pd.to_datetime(df_acc['Colision_hora'], format='%H%M').dt.time
```
